### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

Figure är själva sklettet där man kan rita ut sina "axes" eller "plots", det brukar kallas "canvas". Ett exempel på att skapa en canvas 
= fig, ax = plt.subplots() 

Axes är det område som man skapar inom figure, Axes innehåller alla de visuella komponenterna för att skapa eller kontrollera grafer, titlar, staplar osv. Axes innehåller oftas x- axlar och y-axlar som 2D-grafer men kan också ha y-axel för 3D. 

Axis är enskild axel inom "Axes", med Axis kan man kontrollera "tick märken", det vill säga markera och begränsa genrerade postioner längs axelen. En Axis kan också skriva ut Tick-ettiketer, till exempel vilket namn x-axeln ska visa. Ett exempel kan vara = ax.set_ylabel('Y-axel'), där sätter man en ett namn för "Y-axel". 

Artist är en grundläggande metod som representerar allt som ritas på en Figure. Från en linje eller en text till ett helt koordinatsystem är en typ av Artist. Det finns huvudsäklingen två typer av "Artist", Primära Artist där exempel på det kan vara "Line2D" eller text, till exempel finns det line, = ax.plot([1, 2, 3], [4, 5 ,6]), det är en Line2D och en primitiv Artist. 

kontexstuell Artist används mer för att organisera diagramens struktur och det kan vara "Axes" eller en hel "Figure". Exempel på detta kan vara en: text = ax.text(2, 5 color= blue), kontexstuell Artist.

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
x = np.linspace(-4, 4, 100) # Definerar x värdena med hjälp av "Linspace"

y = x**2 # Definerar funktionen y = x^2 

fix, ax = plt.subplots() # skapar en Figure och axes 

ax.plot(x, y, label=r"$y = x^2$", color = "blue") # Nere till vänster finns en liten ruta där man visar att blå linjen visar y = x2, "dollartecknen" göra så att "2an" kommer ovanför x:et. 
ax.set_xlabel("x") # Namnger x-axlen 
ax.set_xlabel("y") # Namnger y-axlen
ax.set_title("plot of $y = x^2$") # Namnger en Titel 
ax.axhline(0, color="red", linewidth=0.8, linestyle= "--") # Skapar en linje på x-axlen som är röd och har linjestil inte är infylld. 
ax.axvline(0, color="red", linewidth=0.8, linestyle= "--") # Skapar en likadan linje på y-axlen 
ax.grid(alpha=1) # Skapar en rutnät i bakgrunden för att lättare se diagrammet.  
ax.legend() # Skapar en legend.

```

```python
x = np.linspace(-4, 4, 100) # definerar värdena i x-axlen 
y = x**2 # definerar funktionen i y-axlen

plt.plot(x, y, label=r"$y = x^2$", color = "blue")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Plot of y = x^2")
plt.axhline(0, color="red", linewidth=0.8, linestyle="--")
plt.axvline(0, color="red", linewidth=0.8, linestyle="--")
plt.grid(alpha=1)
plt.legend()
```

### Create a figure containing 2  subplots where the first is a scatter plot and the second is a bar plot. You have the data below. 

```python

# Data for scatter plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)
x = np.linspace(-2, 2, 100)
y = x**2

# Data for bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())

fig, (ax1, ax2) = plt.subplots(1, 2, layout='constrained', figsize=(12, 4))

# input for bar plot
plt.bar(names, values, color=['blue'])
plt.title('Fruit count')
plt.xlabel('X')
plt.ylabel('Y')


# input for scatter plot
plt.subplot(1, 2, 1)
plt.scatter(random_data_x, random_data_y, alpha=0.5, color='blue')
plt.title('Plotter scatter')
plt.xlabel('X')
plt.ylabel('y')









```
